```python
from mlxtend.data import loadlocal_mnist
from scipy import stats
from sklearn.metrics import accuracy_score, f1_score

# load training data
X, y = loadlocal_mnist(
    images_path='../emnist/emnist-letters-train-images-idx3-ubyte', 
    labels_path='../emnist/emnist-letters-train-labels-idx1-ubyte')

# load testing data
Xtest, ytest = loadlocal_mnist(
    images_path='../emnist/emnist-letters-test-images-idx3-ubyte', 
    labels_path='../emnist/emnist-letters-test-labels-idx1-ubyte')

baseline = stats.mode(y)

yhat = np.ones_like(ytest)

acc = accuracy_score(ytest, yhat)
f1 = f1_score(ytest, yhat, average='macro')
print(f'Accuracy is {acc}')
print(f'F1 score is {f1}')
```

```python
slice20s = []
slice17s = []
for j in range(W1.shape[1]):
    slice20 = W1[:20,j] 
    slice17 = W1[20:,j]
    slice20s.append(slice20)
    slice17s.append(slice17)
    
slice20s2 = []
slice17s2 = []
for j in range(W2.shape[1]):
    slice20 = W2[:20,j] 
    slice17 = W2[20:,j]
    slice20s2.append(slice20)
    slice17s2.append(slice17)
```

```python
sums20 = [np.sum(vec) for vec in slice20s]
sums17 = [np.sum(vec) for vec in slice17s]
norm20 = [np.linalg.norm(vec) for vec in slice20s]
norm17 = [np.linalg.norm(vec) for vec in slice17s]

sums202 = [np.sum(vec) for vec in slice20s2]
sums172 = [np.sum(vec) for vec in slice17s2]
norm202 = [np.linalg.norm(vec) for vec in slice20s2]
norm172 = [np.linalg.norm(vec) for vec in slice17s2]
```
